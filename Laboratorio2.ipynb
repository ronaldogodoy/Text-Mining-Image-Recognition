#### Problema #1:
Desarrolle una función que permita hacer la binarización de una imagen, para realizarlo puede
utilizar el enfoque que prefiera

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargamos la imagen en escala de grises
img = cv2.imread("Problema1.jpg", cv2.IMREAD_GRAYSCALE)

def binarizar_imagen(img, umbral=127):
    alto = img.shape[0]
    ancho = img.shape[1]
    img_binarizada = np.zeros((alto, ancho), dtype=np.uint8) # Imagen de salida en blanco y negro

    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]
            if pixel >= umbral:
                img_binarizada[i, j] = 255
            else:
                img_binarizada[i, j] = 0

    return img_binarizada
```

```python
# Binarizamos la imagen usando un umbral de 127
img_binarizada = binarizar_imagen(img, umbral=127)
```

```python
# Guardamos y creamos la imagen binarizada
cv2.imwrite("Problema1.binarizada.jpg", img_binarizada)
plt.imshow(img_binarizada, cmap="gray")
plt.title("Imagen Binarizada")
plt.show()
```

#### Problema #2:
Desarrolle una función que dada dos que usted desee del mismo tamaño (igual ancho y alto) y
a color, realice las operaciones SUMA, RESTA, AND, OR , XOR con ambas imágenes recuerde
que para realizar las operaciones lógicas es necesario pasar la imagen a escala de grises y luego
binarizarla.

```python
# Cargue las dos imágenes
img1 = cv2.imread("Problema2.1.jpg")  # Cambia "imagen1.jpg" por el nombre de tu imagen
img2 = cv2.imread("Problema2.2.jpg")  # Cambia "imagen2.jpg" por el nombre de tu segunda imagen

# Convertí a escala de grises las dos imagenes
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# Muestra de las imagenes en escala de grises
plt.subplot(1, 2, 1)
plt.imshow(img1_gray, cmap="gray")
plt.title("Tikal")
plt.subplot(1, 2, 2)
plt.imshow(img2_gray, cmap="gray")
plt.title("Volcan de fuego")
plt.show()
```

#### SUMA

```python
def suma_imagenes(img1, img2):
    suma = cv2.add(img1, img2)
    return suma

suma_resultado = suma_imagenes(img1_gray, img2_gray)
plt.imshow(suma_resultado, cmap="gray")
plt.title("Resultado Suma")
plt.show()
```

### RESTA

```python
def resta_imagenes(img1, img2):
    resta = cv2.subtract(img1, img2)
    return resta

resta_resultado = resta_imagenes(img1_gray, img2_gray)
plt.imshow(resta_resultado, cmap="gray")
plt.title("Resultado Resta")
plt.show()
```

### AND

```python
def and_imagenes(img1, img2):
    and_op = cv2.bitwise_and(img1, img2)
    return and_op

and_resultado = and_imagenes(img1_gray, img2_gray)
plt.imshow(and_resultado, cmap="gray")
plt.title("Resultado AND")
plt.show()
```

### OR

```python
def or_imagenes(img1, img2):
    or_op = cv2.bitwise_or(img1, img2)
    return or_op

or_resultado = or_imagenes(img1_gray, img2_gray)
plt.imshow(or_resultado, cmap="gray")
plt.title("Resultado OR")
plt.show()
```

### XOR

```python
def xor_imagenes(img1, img2):
    xor_op = cv2.bitwise_xor(img1, img2)
    return xor_op

xor_resultado = xor_imagenes(img1_gray, img2_gray)
plt.imshow(xor_resultado, cmap="gray")
plt.title("Resultado XOR")
plt.show()
```

#### Problema #3:
Desarrolle una función que dada una imagen en escala de grises muestre dicha imagen en 3 dimensiones, donde los ejes X e Y representan el dominio espacial de la imagen y Z representa la intensidad del color gris de cada pixel.


```python
from mpl_toolkits.mplot3d import Axes3D

# Cargamos la imagen en escala de grises
img = cv2.imread("Problema3.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
def imagen_3d(img):
    # Creación de la malla de coordenadas X, Y
    alto, ancho = img.shape
    X, Y = np.meshgrid(np.arange(0, ancho), np.arange(0, alto))
    
    # Intensidad de la imagen en Z
    Z = img

    # Creación de la figura y los ejes 3D
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')

    # Grafiqué la superficie en 3d
    ax.plot_surface(X, Y, Z, cmap='gray')

    # Etiquetas asignados a los ejes
    ax.set_xlabel('Eje X')
    ax.set_ylabel('Eje Y')
    ax.set_zlabel('Intensidad de Color')

    # Muestra de la grafica en 3d
    plt.show()
```

```python
imagen_3d(img)
```

#### Problema #4:
Implemente una función para cada una de las transformaciones vistas en clase: negativa, lineal, logarítmica y exponencial.

```python
# Cargamos la imagen en escala de grises
img = cv2.imread("Problema4.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# Imagen original
plt.imshow(img, cmap='gray')
plt.title("Imagen Original")
plt.axis('off')
plt.show()
```

### Transformación Negativa

```python
def transformar_negativa(img):
    # Transformación negativa: pixel = 255 - pixel
    img_negativa = 255 - img
    return img_negativa

# Cargamos la imagen en escala de grises
img = cv2.imread("Problema4.jpg", cv2.IMREAD_GRAYSCALE)

# Aplicamos la transformación negativa
img_negativa = transformar_negativa(img)

# Guardamos y mostramos el resultado
cv2.imwrite("Problema4_negativa.jpg", img_negativa)

plt.imshow(img_negativa, cmap='gray')
plt.title("Transformación Negativa")
plt.axis('off')
plt.show()
```

### Transformación Lineal

```python

def transformar_lineal(img, alpha=1.0, beta=0):
    # Transformación lineal: pixel = alpha * pixel + beta
    img_lineal = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
    return img_lineal

# Cargamos la imagen en escala de grises
img = cv2.imread("Problema4.jpg", cv2.IMREAD_GRAYSCALE)

# Aplicamos la transformación lineal
img_lineal = transformar_lineal(img, alpha=1.2, beta=50)  # Ajusta los parámetros según sea necesario

# Guardamos y mostramos el resultado
cv2.imwrite("Problema4_lineal.jpg", img_lineal)

plt.imshow(img_lineal, cmap='gray')
plt.title("Transformación Lineal")
plt.axis('off')
plt.show()
```

### Transformación Logarítmica

```python
def transformar_logaritmica(img):
    # Transformación logarítmica: pixel = c * log(1 + pixel)
    c = 255 / np.log(1 + np.max(img))
    img_logaritmica = c * np.log(1 + img)
    img_logaritmica = np.uint8(img_logaritmica)
    return img_logaritmica

# Cargamos la imagen en escala de grises
img = cv2.imread("Problema4.jpg", cv2.IMREAD_GRAYSCALE)

# Aplicamos la transformación logarítmica
img_logaritmica = transformar_logaritmica(img)

# Guardamos y mostramos el resultado
cv2.imwrite("Problema4_logaritmica.jpg", img_logaritmica)

plt.imshow(img_logaritmica, cmap='gray')
plt.title("Transformación Logarítmica")
plt.axis('off')
plt.show()
```

### Transformación Exponencial

```python
def transformar_exponencial(img):
    # Transformación exponencial: pixel = c * (exp(pixel / c) - 1)
    c = 255 / (np.exp(1) - 1)
    img_exponencial = c * (np.exp(img / c) - 1)
    img_exponencial = np.uint8(img_exponencial)
    return img_exponencial

# Cargamos la imagen en escala de grises
img = cv2.imread("Problema4.jpg", cv2.IMREAD_GRAYSCALE)

# Aplicamos la transformación exponencial
img_exponencial = transformar_exponencial(img)

# Guardamos y mostramos el resultado
cv2.imwrite("Problema4_exponencial.jpg", img_exponencial)

plt.imshow(img_exponencial, cmap='gray')
plt.title("Transformación Exponencial")
plt.axis('off')
plt.show()
```

#### Problema #5:
Investigue que hace el kernel Laplaceano y como este enfoque se diferencia de usar el algoritmo de Canny para la detección de bordes

### Kernel Laplaciano

El kernel laplaciano es un operador diferencial que, en el procesamiento de imágenes, se utiliza principalmente para detectar cambios bruscos en la intensidad de la imagen. Estos cambios bruscos son, en esencia, los bordes.

El kernel laplaciano se aplica como un filtro a la imagen. Este filtro, en su forma más simple, es una matriz de números que se convoluciona con la imagen. Los valores altos en la salida de esta convolución indican la presencia de un borde.

Al aplicar el laplaciano, estamos calculando la segunda derivada de la intensidad de la imagen. En los bordes, la intensidad cambia rápidamente, lo que resulta en valores altos de la segunda derivada.

**Ventajas**

> - **Sencillez:** Es un operador computacionalmente eficiente.
> - **Robustez:** Funciona bien en una variedad de tipos de imágenes.

**Desventajas**
> - **Sensibilidad al ruido:** El laplaciano amplifica el ruido presente en la imagen, lo que puede llevar a la detección de falsos bordes.
> - **Doble respuesta:** El laplaciano responde tanto al borde interior como al exterior, lo que puede dificultar la localización precisa del borde.

### ¿Cómo se diferencia de Canny?

El algoritmo de Canny es un enfoque más sofisticado para la detección de bordes que aborda algunas de las limitaciones del laplaciano.

* **Múltiples etapas:** Canny involucra múltiples etapas, incluyendo:

> - **Reducción de ruido:** Emplea un filtro gaussiano para suavizar la imagen y reducir el ruido antes de la detección de bordes.
> - **Cálculo del gradiente:** Calcula la magnitud y la dirección del gradiente de la imagen para determinar la ubicación y la orientación de los bordes.
> - **Supresión no máxima:** Elimina los píxeles que no corresponden a los máximos locales del gradiente, asegurando que solo se detecten bordes finos.
> - **Umbralización:** Utiliza dos umbrales para determinar qué bordes son fuertes, débiles o deben ser suprimidos.
> - **Histeresis:** Conecta los bordes débiles a los fuertes para formar contornos completos

**Ventajas de Canny:**

> - **Precisión:** Proporciona una detección de bordes más precisa y robusta.
> - **Completitud:** Detecta bordes completos y bien definidos.
> - **Adaptabilidad:** Puede ser ajustado para diferentes tipos de imágenes y niveles de ruido.
